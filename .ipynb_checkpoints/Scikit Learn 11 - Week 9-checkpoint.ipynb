### Week  9 - Mengenal TF-IDF (Term Frequency - Inverse Document Frequency)

1. Dataset

TF-IDF (Term Frequency - Inverse Document Frequency) adalah salah satu metode statistik yang digunakan untuk mengukur seberapa penting suatu kata terhadap sebuah dokumen tertentu dari sekumpulan dokumen atau corpus. TF-IDF pada umumnya melibatkan perkalian antara dia nilai yaitu nilai TF dan nilai IDF.   Berikut link referensi yang bisa digunakan:
- https://en.wikipedia.org/wiki/Tf%E2%80%93idf
- https://scikit-learn.org/stable/modules/feature_extraction.html#text-feature-extraction

- Berikut beberapa opsi formula yang bisa digunakan untuk merepresentasikan Term Frequency:



Dalam Scikit Learn, formula Term Frequency yang digunakan adalah hasil pembagian dari jumlah kemunculan suatu term pada dokumen dengan total jumlah kata pada dokumen tersebut.

- Berikut beberapa opsi formula yang bisa digunakan untuk merepresentasikan Inverse Document Frequency:



Dalam Scikit Learn, formula Inverse Document Frequency yang digunakan adalah:



$n$ = jumlah dokumen dalam suatu corpus

df$(t)$ = jumlah dokumen yang mengandung term tertentu
