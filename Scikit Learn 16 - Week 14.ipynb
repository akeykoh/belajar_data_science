#### General ML Model Training



Dalam proses training, pasti diawali dengan training set dimana didalamnya terdiri dari sekumpulan training features dan sekumpulan target label. Training set akan digunakan untuk melakukan training model machine learning. Training model berguna untuk memprediksi sekumpulan nilai feature baru. 

#### Ensemble Learning: Heterogeneous & Homogeneous



Ensemble learning merupakan teknik machine learning dimana kita menggabungkan beberapa model untuk melakukan prediksi. Ensemble learning juga diawali dengan suatu training set. Training set ini akan digunakan untuk melakukan training beberapa model. Contohnya pada kasus ini adalah KNN, SVM, dan Decision Tree.

Proses training akan menghasilkan 3 trained model. Setiap trained model akan digunakan untuk melakukan prediksi terhadap sekumpulan nilai features yang baru. Setiap hasil prediksi ini akan disatukan dengan mekanisme majority voting. 

Untuk kasus regression task, majority voting dilakukan dengan menggunakan nilai mean / rata-rata. Untuk kasus classification task, majority voting dilakukan dengan menerapkan mode / kemunculan terbanyak. Untuk contoh diatas, hasil prediksi final akan direpresentasikan sebagai `y_pred` final. 

Heterogeneous merupakan model ensemble learning yang heterogen (memadukan sejumlah model machine learning yang berbeda). Homogeneous merupakan model ensemble learning yang homogen (memadukan sejumlah model machine learning yang sejenis). 

Berikut link referensi yang bisa digunakan untuk pembelajaran lebih lanjut: https://en.wikipedia.org/wiki/Ensemble_learning

#### Bagging: Bootstrap Aggregating



Gambar diatas merupakan salah satu contoh untuk homogeneous ensemble learning. Dikarenakan kita menerapkan sejumlah model yang sejenis, maka melakukan training untuk setiap model dengan training set yang sama akan sia-sia saja. Ini dikarenakan trained model yang dihasilkan akan sama persis. Salah satu teknik yang bisa digunakan untuk mengatasi masalah ini adalah bootstrap aggregating / bagging. 

Bagging akan menerapkan random sampling with replacement terhadap training set yang dimiliki, setelah itu akan dihasilkan beberapa training set baru yang akan di training. Training set yang dihasilkan biasanya dinamakan bag. Bag berisi sekumpulan sample data. Setiap isi bag berbeda-beda. 

Bag akan digunakan untuk mentraining 3 model machine learning yang sejenis. Setelah itu, akan dihasilkan trained model. Setiap trained model akan digunakan untuk memprediksi terhadap sekumpulan nilai features yang baru. Hasil prediksi akan disatukan menggunakan proses majority voting dan akan dihasilkan hasil prediksi final. 

Berikut link referensi yang bisa digunakan untuk pembelajaran lebih lanjut:
https://en.wikipedia.org/wiki/Bootstrap_aggregating

#### Random Forest



Random Forest merupakan implementasi homogeneous ensemble learning yang menerapkan decision tree. Untuk contoh diatas, model yang akan di training adalah decision tree. Random forest dibentuk dari sekumpulan decision tree. Random forest menerapkan bagging dan features randomness, dimana setiap bag yang dihasilkan akan mengadopsi sejumlah features yang dipilih secara acak dari training set sumbernya. 

Dikarenakan decision tree ditraining menggunakan data set yang berbeda, maka akan dihasilkan beberapa trained model. Trained model ini akan digunakan untuk melakukan prediksi terhadap sekumpulan nilai features yang baru. Hasil prediksi akan disatukan menggunakan proses majority voting dan akan dihasilkan hasil prediksi final. Random forest memiliki performa yang lebih superior daripada model decision tree yang biasa. 

#### Dataset

Untuk mempelajari implementasi random forest dengan menggunakan Scikit Learn, kita akan menyiapkan datasetnya terlebih dahulu. Dataset yang digunakan adalah iris dataset. Berikut link untuk mempelajari lebih lanjut mengenai iris dataset: 
https://en.wikipedia.org/wiki/Iris_flower_data_set

```python
# import modul load_iris
from sklearn.datasets import load_iris

# memanggil load_iris dan menyertakan parameter
X, y = load_iris(return_X_y=True)

print(f'Dimensi Feature: {X.shape}')
print(f'Class: {set(y)}')
```

```python
# import modul train_test_split
from sklearn.model_selection import train_test_split

# memanggil train_test_split dan menyertakan parameter
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=0)
```

#### Classification dengan `RandomForestClassifier`

Kita akan melakukan classification spesies iris dengan menggunakan `RandomForestClassifier`.

```python
# import modul RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

# membentuk object dari RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, # banyak decision tree yang akan ditraining
                               random_state=0)   # agar bisa di replicate dengan hasil yang konsisten

# melakukan training menggunakan metode fit
model.fit(X_train, y_train)
```
