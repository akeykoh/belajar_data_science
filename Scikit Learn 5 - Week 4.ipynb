#### Sample Dataset

Pada contoh ini, kita akan membentuk dataset dengan format pandas dataframe yang berisi daftar dari diameter pizza dan harganya.

```python
# mengimport modul pandas
import pandas as pd

# menyiapkan dictionary dimana setiap keynya akan berasosiasi dengan list
pizza = {'diameter': [6, 8, 10, 14, 18],
         'harga': [7, 9, 13, 17.5, 18]}

# membentuk object dataframe dari dictionary yang sudah dibentuk
pizza_df = pd.DataFrame(pizza)

# menampung object dataframe yang dibentuk dalam variable pizza_df
pizza_df
```

#### Visualisasi Data

Pada contoh ini, kita akan melakukan visualisasi dataset yang sudah kita bentuk menjadi scatter plot. Sumbu x akan berasosiasi dengan diameter. Sumbu y akan berasosiasi dengan harga. Untuk melakukan visualisasi dataset menjadi scatter plot, kita bisa memanggil function `.plot`. Function ini membutuhkan 3 parameter yaitu parameter `kind` dengan nilai `scatter`, parameter sumbu x, dan parameter sumbu y.

```python
# mengimport modul matplotlib.pyplot
import matplotlib.pyplot as plt

# melakukan visualisasi data menjadi scatter plot
pizza_df.plot(kind='scatter', x='diameter', y='harga')

# melakukan pengaturan title dan label
plt.title('Perbandingan Diameter dan Harga Pizza')
plt.xlabel('Diameter (inch)')
plt.ylabel('Harga (dolar)')

# melakukan pengaturan min dan max untuk x dan y axis
plt.xlim(0, 25)
plt.ylim(0, 25)

# mengaktifkan grid
plt.grid(True)

# menunjukkan hasil scatter plot
plt.show()
```

#### Penyesuaian Dataset

Pada contoh dibawah kita akan melakukan beberapa penyesuaian untuk dataset yang sudah kita bentuk. Kita akan mengelompokkan features dan target ke dalam 2 variable terpisah. Kita juga akan mengonversi struktur data yang awalnya pandas dataframe, menjadi numpy array. 

```python
# mengimport modul numpy
import numpy as np

# melakukan konversi menjadi numpy array
X = np.array(pizza_df['diameter']) # variable feature
y = np.array(pizza_df['harga'])    # variable target

print(f'X: {X}')
print(f'y: {y}')
```

Dari output diatas, kita bisa melihat bahwa featuresnya (variable X) hanya terdiri dari 1 dimensi. Seharusnya model machine learning dalam Scikit Learn memerlukan sekumpulan nilai features yang ditampung dalam suatu array 2 dimensi. Oleh karena itu, kita akan melakukan reshaping terhadap array yang menampung nilai features. 

```python
# melakukan reshaping untuk array yang menampung nilai features
X = X.reshape(-1, 1)

# menampilkan dimensi dari variable X
X.shape
```

```python
# mengecek nilai dalam variable X
X
```

#### Training Simple Linear Regression Model

Kita bisa melakukan training model machine learning dengan model `LinearRegression`. 

```python
# mengimport LinearRegression
from sklearn.linear_model import LinearRegression

# membuat object model dari estimator class
model = LinearRegression()

# object model akan di training dengan method fit
model.fit(X, y)
```

#### Visualisasi Simple Linear Regression Model

Untuk dapat memvisualisasikan model ini, kita harus menyiapkan 2 buah nilai untuk features. Pada contoh dibawah 2 nilai featuresnya adalah 0 dan 25. Ini artinya kita akan melakukan plotting pada area dengan ukuran 25 x 25. Nilai terkecilnya adalah 0 dan nilai terbesarnya adalah 25. 

```python
# menyiapkan 2 buah nilai untuk features dan dibentuk menjadi numpy array
# setelah itu akan dilakukan reshaping supaya menghasilkan array 2 dimensi
X_vis = np.array([0, 25]).reshape(-1, 1)

# melakukan prediksi nilai y berdasarkan nilai X_vis 
y_vis = model.predict(X_vis)
```

```python
# melakukan plotting marker warna biru
plt.scatter(X, y) 

# melakukan plotting garis merah
plt.plot(X_vis, y_vis, '-r')

# melakukan pengaturan title dan label
plt.title('Perbandingan Diameter dan Harga Pizza')
plt.xlabel('Diameter (inch)')
plt.ylabel('Harga (dollar)')

# melakukan pengaturan min dan max untuk x dan y axis
plt.xlim(0,25)
plt.ylim(0,25)

# mengaktifkan grid
plt.grid(True)

# menunjukkan hasil visualisasi
plt.show()
```

Garis linear yang terbentuk akan mempunyai formula persamaan yang mengacu pada formula linear regression.

Formula Linear Regression:  =  + x

- : response variable / target
- : explanatory variable / features
- : intercept (nilai pada sumbu y ketika sumbu xnya bernilai nol)
- : slope (slope akan berpengaruh tingkat kemiringan dari garis linear. Untuk nilai slope nol, maka akan terbentuk garis horizontal)

Kita juga bisa mengetahui nilai intercept dan nilai slope dari model linear regression yang sudah kita lakukan training. Caranya seperti dibawah ini.

```python
# mengetahui nilai intercept
print(f'intercept: {model.intercept_}')

# mengetahui nilai slope
print(f'slope: {model.coef_}')
```
