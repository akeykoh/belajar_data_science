#### Formula Dasar Logistic Regression

Logistic Regression digunakan untuk melakukan prediksi nilai yang berada pada rentang nilai antara 0 dan 1. Penerapan logistic regression pada kasus binary classification task adalah dengan meng endcode kedua kelas dalam binary classification sebagai 0 (jika probability mendekati 0) dan 1 (jika probability mendekati 1). Berikut formula dasar dari logistic regression:

- Simple Linear Regression (hanya bisa menyatakan 1 feature)

$y = \alpha + \beta x$

$g(x) = \alpha + \beta x$

- Multiple Linear Regression (menyatakan lebih dari 1 feature)

$y = \alpha + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_n x_n$

$g(X) = \alpha + \beta X$

- Logistic Regression

$g(X) = sigmoid(\alpha + \beta X)$

$sigmoid(x) = \frac {1} {1 + exp(-x)}$



Kurva sigmoid memiliki bentuk seperti huruf S. Rentang nilai sumbu X tergantung dari kasus yang dihadapi. Rentang nilai sumbu y pada kurva sigmoid adalah 0 dan 1. Rentang nilai ini akan berkolerasi dengan binomial probability. 

#### Dataset: SMS Spam Collection Data Set

Berikut link dari sumber data yang digunakan: https://archive.ics.uci.edu/ml/datasets/SMS+Spam+Collection

```python
# mengimport modul pandas
import pandas as pd 

# meload file sebagai pandas data frame
df = pd.read_csv('./data/SMSSpamCollection', # posisi dari data set
                sep='\t',                    # separator tab
                header = None,               # tidak ada header
                names = ['label', 'sms'])    # nama kolom

# mencetak 5 baris pertama
df.head()
```

```python
# mengitung banyak data ham dan spam
df['label'].value_counts()
```

#### Training and Testing Dataset

Kita akan membagi dataset yang kita gunakan menjadi 2 bagian. Kita akan mengelompokkan features dan target terlebih dahulu. 

```python
# import modul LabelBinarizer
from sklearn.preprocessing import LabelBinarizer

X = df['sms'].values   # menampung nilai dari kolom sms
y = df['label'].values # menampung nilai dari kolom label

# membentuk object dari class LabelBinarizer
lb = LabelBinarizer() 

# melakukan fit transform dari object LabelBinarizer yang akan dikenakan ke nilai variabel y
y = lb.fit_transform(y).ravel()

# 0 = ham
# 1 = spam
lb.classes_
```

```python
# import modul train_test_split
from sklearn.model_selection import train_test_split

# memanggil train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.25, # porsi testing dataset
                                                    random_state=0) 
# menampilkan X_train dan y_train
print(X_train, '\n')
print(y_train)
```

#### Feature Extraction dengan TF-IDF

Kita akan menerapkan TF-IDF pada data set kita. 

```python
# import modul TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

# membentuk object TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')

# memanggil function TfidfVectorizer
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

print(X_train_tfidf)
```

#### Binary Classification dengan Logistic Regression

Untuk mengklasifikasi ham dan spam pada dataset SMS, kita akan menggunakan Logistic Regression 

```python
# import modul LogisticRegression
from sklearn.linear_model import LogisticRegression

# membentuk object class Logistic Regression
model = LogisticRegression()

# melakukan training object class Logistic Regression
model.fit(X_train_tfidf,y_train)

# melakukan prediksi label berdasarkan data-data features
y_pred = model.predict(X_test_tfidf)

# menampilkan 5 hasil prediksi pertama
for pred, sms in zip(y_pred[:5], X_test[:5]):
    print(f'PRED: {pred} - SMS: {sms}\n')
```

#### Evaluation Metrics pada Binary Classification

Beberapa Evaluation Metrics yang bisa kita gunakan untuk kasus Binary Classification: 
- Confussion matrix
- Accuracy
- Precision and Recall
- F1 score
- ROC

#### Terminologi Dasar Evaluation Metrics

Merepresentasikan hasil klasifikasi / prediksi yang benar:
- True Positive (TP) adalah sesuatu yang bernilai positif dan diprediksi benar sebagai nilai positif oleh model.
- True Negative (TN) adalah sesuatu yang bernilai negatif dan diprediksi benar sebagai nilai negatif oleh model.

Merepresentasikan hasil klasifikasi / prediksi yang salah:
- False Positive (FP) merupakan sesuatu yang bernilai negatif dan diprediksi keliru sebagai nilai positif oleh model.
- False Negative (FN) merupakan sesuatu yang bernilai positif dan diprediksi keliru sebagai nilai negatif oleh model.

#### Confusion Matrix 

Biasanya Confusion Matrix disebut juga dengan error matrix. 

Berikut link referensi untuk mempelajari Confusion Matrix lebih lanjut: https://en.wikipedia.org/wiki/Confusion_matrix

```python
# import modul confusion_matrix
from sklearn.metrics import confusion_matrix

# memanggil function confusion_matrix
matrix  = confusion_matrix(y_test, y_pred)
matrix
```

```python
# mengkonversikan menjadi array 1 dimensi
tn, fp, fn, tp = matrix.ravel()

print(f'TN: {tn}')
print(f'FP: {fp}')
print(f'FN: {fn}')
print(f'TP: {tp}')
```

```python
# import modul matplotlib.pyplot
import matplotlib.pyplot as plt

plt.matshow(matrix)
plt.colorbar() 

# mengatur title dan label
plt.title('Confusion Matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```
