#### Bayes' Theorem

Bayes' theorem menawarkan suatu formula untuk menghitung nilai probability dari suatu event dengan memanfaatkan pengetahuan sebelumnya dari kondisi terkait; atau biasanya dikenal dengan istilah conditional probability.

Formula untuk mengekspresikan probability dari kemunculan event A, apabila event B muncul: 

$P(A|B) = \frac {P(B|A) \times P(A)} {P(B)}$

Formula probability dari kemunculan nilai target label y, apabila diketahui kemunculan nilai feature X. Ini merupakan penyesuaian untuk konteks pengklasifikasian:

$P(y|X) = \frac {P(X|y) \times P(y)} {P(X)}$


Formula jika mengadopsi terminology Naive Bayes:

$Posterior = \frac {Likelihood} {Evidence}$

#### Pengenalan Naive Bayes Classification

##### Studi Kasus 1



Asep dan Joko akan berperan sebagai target label atau class.

Untuk studi kasus ini, kita memiliki misi untuk melakukan prediksi siapa pelanggan yang akan melakukan pemesanan dengan diketahui pesanannya adalah lumpia dan bakso. Berikut beberapa terminologi dasar dalam Naive Bayes: 

 - Prior Probability: $P(y)$

Prior Probability merupakan nilai probability dari kemunculan suatu nilai target label tertentu tanpa memperhatikan nilai featurenya. Untuk kasus ini, kita bisa mengasumsikan bahwa peluang untuk suatu pemesanan yang dilakukan oleh Asep atau Joko berimbang yaitu 0.5 atau 50%.

$P(Asep) = 0.5$

$P(Joko) = 0.5$

Berikut link referensi untuk mempelajari lebih lanjut mengenai Prior Probability:
https://en.wikipedia.org/wiki/Prior_probability

- Likelihood: $P(X|y)$

Likelihood merupakan probability kemunculan nilai feature tertentu jika diketahui kemunculan nilai target labelnya. Kita bisa menggambarkan likelihood pada studi kasus 1 seperti berikut:
1. Asep:

$P(lumpia, bakso|Asep) = (0.1 \times 0.8) = 0.08$

2. Joko:

$P(lumpia, bakso|Joko) = (0.3 \times 0.2) = 0.06$

Berikut link referensi untuk mempelajari lebih lanjut mengenai Likelihood:
https://en.wikipedia.org/wiki/Likelihood_function

- Evidence atau Normalizer: $P(X)$

Evidence merupakan total akumulasi dari hasil perkalian antara nilai Likelihood dan nilai Prior. Formula dari Evidence adalah:

$Evidence = \sum (Likelihood \times Prior$

Kita bisa menggambarkan Evidence pada studi kasus 1 seperti berikut:

$P(lumpia, bakso) = (0.08 \times 0.5) + (0.06 \times 0.5) = 0.07$

- Posterior Probability: $P(y|X)$

Posterior Probability adalah nilai probability kemunculan suatu class atau target label dengan diketahui kemunculan sekumpulan nilai featurenya. Formula Posterior adalah:

$Posterior = \frac {Likelihood \times Prior} {Evidence}$

Kita bisa menggambarkan Posterior Probability pada studi kasus 1 seperti berikut:

$P(Asep|lumpia, bakso) = \frac {0.08 \times 0.5} {0.07} = 0.57$

$P(Joko|lumpia, bakso) = \frac {0.06 \times 0.5} {0.07} = 0.43$

Berikut link referensi untuk mempelajari lebih lanjut mengenai Posterior Probability:
https://en.wikipedia.org/wiki/Posterior_probability

Bisa kita lihat bahwa nilai posterior Asep lebih tinggi dibandingkan Joko. Oleh karena itu Naive Bayes akan mengklasifikasikan Asep sebagai pemesannya.

##### Studi Kasus 2



Asep dan Joko akan berperan sebagai target label atau class.

Untuk studi kasus ini, kita memiliki misi untuk melakukan prediksi siapa pelanggan yang akan melakukan pemesanan dengan diketahui pesanannya adalah siomay dan bakso. Berikut beberapa terminologi dasar dalam Naive Bayes: 

- Prior Probability: $P(y)$

$P(Asep) = 0.5$

$P(Joko) = 0.5$

- Likelihood: $P(X|y)$
1. Asep:

$P(lumpia, bakso|Asep) = (0.1 \times 0.8) = 0.08$

2. Joko:

$P(lumpia, bakso|Joko) = (0.5 \times 0.2) = 0.1$

- Evidence atau Normalizer: $P(X)$ 

$P(siomay, bakso) = (0.1 \times 0.8 \times 0.5) + (0.5 \times 0.2 \times 0.5) = 0.09$

- Posterior Probability: $P(y|X)$

$P(Asep|siomay, bakso) = \frac {(0.1 \times 0.8) \times 0.5} {0.09} = 0.444$

$P(Joko|siomay, bakso) = \frac {(0.5 \times 0.2) \times 0.5} {0.09} = 0.555$

Bisa kita lihat bahwa nilai posterior Joko lebih tinggi dibandingkan Asep. Oleh karena itu Naive Bayes akan mengklasifikasikan Joko sebagai pemesannya.
